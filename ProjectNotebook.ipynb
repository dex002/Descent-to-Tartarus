# Descent to Tartarus

This project is a turn-based dungeon crawler based off the classic roguelike dungeon crawler genre. The player selects a party of two heros from four unique classes each with their own unique skills. 

## Classes:
- Aphros — The Healer. Based on Aphrodite. Heals allies, deals small damage, and reduces enemy attack.
- Aspirda — The Tank. Based on Hephaestus. Increases own and ally defense, and slows enemies.
- Doros — The Assassin. Based on Ares. Deals massive damage, steals HP, and builds defense over time.
- Lusios — The Madman. Based on Dionysus. Builds hysteria stacks for scaling damage and has random chance-based skills.

## Gameplay:
Your party will enter dungeons, fighting randomly generated enemies that scale in difficulty with each floor. Allies and enemies will attack in an order based off of an inititive roll that is influenced by the combatants speed stat. 

The party then enters endless dungeons, fighting randomly generated enemies that scale in difficulty with each floor. The game ends when both heroes are defeated, and the goal is simply to survive as many floors as possible.
